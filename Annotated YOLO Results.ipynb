```python
from ultralytics import YOLO

# Specify the full path to your data YAML file
yaml_filename = '/kaggle/.................../data.yaml'
# This line specifies the full path to the YAML configuration file (data.yaml) for the YOLO model. The YAML file typically contains configuration settings such as the dataset directory, class names, image sizes, etc.
# Create YOLO model
model = YOLO('yolov8n.yaml')  # Creates a YOLO object from 'yolov8n.yaml' configuration file.
#This configuration file contains the architecture and hyperparameters of the YOLOv8 model.
model = model.load('yolov8n.pt')  # Loads pretrained weights

# Train the model
model.train(data=f'{yaml_filename}', epochs=100)
```
